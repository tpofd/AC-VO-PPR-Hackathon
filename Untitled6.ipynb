```
!pip install pandas
!pip install catboost
import pandas as pd
import xgboost as xgb
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
from collections import Counter
from tqdm import tqdm
from sklearn.cluster import KMeans
```

```
train_X = train.copy().drop("class", axis=1)
train_y = train.copy()["class"]
```

```
test_X = test.copy().drop("class", axis=1)
test_y = test.copy()["class"]
```

```
from sklearn import preprocessing
le_protocol = preprocessing.LabelEncoder()
le_protocol.fit(train_X["protocol_type"])
train_X["protocol_type"] = le_protocol.transform(train_X["protocol_type"])
test_X["protocol_type"] = le_protocol.transform(test_X["protocol_type"])

le_service = preprocessing.LabelEncoder()
le_service.fit(train_X["service"])
train_X["service"] = le_service.transform(train_X["service"])
test_X["service"] = le_service.transform(test_X["service"])

le_flag = preprocessing.LabelEncoder()
le_flag.fit(train_X["flag"])
train_X["flag"] = le_flag.transform(train_X["flag"])
test_X["flag"] = le_flag.transform(test_X["flag"])


le_label = preprocessing.LabelEncoder()
le_label.fit(train_y)
train_y = le_label.transform(train_y)
test_y = le_label.transform(test_y)
```

```
train_X.head()
```

```
train_y
```

```
cat = CatBoostClassifier(iterations = 2000, learning_rate = 0.05, depth = 3, loss_function = 'CrossEntropy', eval_metric='Accuracy',
                         random_state = 53, use_best_model=True)

train_data = Pool(
    data = train_X,
    label = train_y
)

valid_data = Pool(
    data = test_X,
    label = test_y
)
```

```
cat.fit(train_data, eval_set = valid_data)
```

```
train_multi = pd.read_csv("/content/drive/My Drive/KDD/KDDTrain+_meted.csv")
test_multi = pd.read_csv("/content/drive/My Drive/KDD/KDDTest+_meted.csv")
```

```
test_multi.head()
```

```
train_multi = train_multi.drop("Unnamed: 42", axis=1)
test_multi = test_multi.drop("Unnamed: 42", axis=1)
```

```
train_multi.shape
```

```
test_unknown = ['apache2', 'httptunnel', 'mailbomb', 'mscan', 'named', 'processtable', 'ps', 'saint',
'sendmail', 'snmpgetattack', 'snmpguess', 'sqlattack', 'udpstorm', 'worm', 'xlock', 'xsnoop', 'xterm']

train_multi = train_multi[train_multi['class'] != "normal"]
test_multi = test_multi[test_multi['class'] != "normal"]

for name in test_unknown:
  test_multi = test_multi[test_multi['class'] != name]
```

```
train_multi_X = train_multi.copy().drop("class", axis=1)
train_multi_y = train_multi.copy()["class"]

test_multi_X = test_multi.copy().drop("class", axis=1)
test_multi_y = test_multi.copy()["class"]
```

```
set(test_multi_y) - set(train_multi_y)
```

```
le_protocol_multi = preprocessing.LabelEncoder()
le_protocol_multi.fit(train_multi_X["protocol_type"])
train_multi_X["protocol_type"] = le_protocol_multi.transform(train_multi_X["protocol_type"])
test_multi_X["protocol_type"] = le_protocol_multi.transform(test_multi_X["protocol_type"])

le_service_multi = preprocessing.LabelEncoder()
le_service_multi.fit(train_multi_X["service"])
train_multi_X["service"] = le_service_multi.transform(train_multi_X["service"])
test_multi_X["service"] = le_service_multi.transform(test_multi_X["service"])

le_flag_multi = preprocessing.LabelEncoder()
le_flag_multi.fit(train_multi_X["flag"])
train_multi_X["flag"] = le_flag_multi.transform(train_multi_X["flag"])
test_multi_X["flag"] = le_flag_multi.transform(test_multi_X["flag"])


le_label_multi = preprocessing.LabelEncoder()
le_label_multi.fit(train_multi_y)
train_multi_y = le_label_multi.transform(train_multi_y)
test_multi_y = le_label_multi.transform(test_multi_y)
```

```
cat_multi = CatBoostClassifier(iterations = 1000, learning_rate = 0.1, depth = 5, loss_function = 'MultiClass', eval_metric='Accuracy',
                         random_state = 53, use_best_model=True)

train_data_multi = Pool(
    data = train_multi_X,
    label = train_multi_y
)

valid_data_multi = Pool(
    data = test_multi_X,
    label = test_multi_y
)
```

```
cat_multi.fit(train_data_multi, eval_set = valid_data_multi)
```
